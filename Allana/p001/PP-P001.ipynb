````python
a = float(5.5)
b = float(12.3)

#Operadores aritmeticos
c = a + b
print("c = a + b: ", c, " = ", a, " + ", b)
c = a - b
print("c = a - b: ", c, " = ", a, " - ", b)
c = a * b
print("c = a * b: ", c, " = ", a, " * ", b)
c = a / b
print("c = a / b: ", c, " = ", a, " / ", b) 
c = a // b
print("c = a // b: ", c, " = ", a, " // ", b) 
c = a % b
print("c = a '%' b: ", c, " = ", a, " % ", b) 
c = a ** b
print("c = a ** b: ", c, " = ", a, " ** ", b) 
c = -a
print("c = -a: ", c, " = -", a) 

#Operadores aritmeticos compostos
c += b
print("c += b: ", c, " = ", a, " + ", b)
c -= b
print("c -= b: ", c, " = ", a, " - ", b) 
c *= b
print("c *= b: ", c, " = ", a, " * ", b) 
c /= b
print("c /= b: ", c, " = ", a, " / ", b) 
c //= b
print("c //= b: ", c, " = ", a, " // ", b) 
c %= b
print("c %= b: ", c, " = ", a, " % ", b) 
c **= b
print("c **= b: ", c, " = ", a, " ** ", b) 

vp = float(2)
vn = float(2)
while True:
    try:
        vp **= 2
        vn **= -2
    except:
        break
print(vp)
print(vn)
        

#Variável numérica é imutável:
#Considerando que  b = a, ambos devem ser iguais a 5.5
b = a
print("a = ", a, " b = ", b)

#Alterar uma das variáveis deveria alterar a outra já que ambas apontam para o mesmo espaço da memória
#Porém, como são variáveis numéricas isso não ocorrerá

a += 1.1
print("a = ", a, " b = ", b)

print("Os métodos disponíveis para variáveis de ponto flutuante são:")
print(dir(float))
````
