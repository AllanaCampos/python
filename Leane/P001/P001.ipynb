```python
a = 10
b = 3

#Operadores aritmeticos
print('Operadores aritmeticos: ')
#Soma 
x = a + b
print('soma_x = ', x)
#Subtração
x = a - b 
print('subtração_x = ', x)
#Multiplicação
x = a * b 
print('multiplicação_x = ', x)
#Divisão
x = a / b 
print('divisão_x = ', x)
#Divisão que retorna apenas a parte inteira
x = a // b 
print('divisão_x = ', x)
#Módulo
x = a % b 
print('módulo_x = ', x)
#Potencia
x = a * b 
print('potencia_x = ', x)
#Numero negativo
x = -a 
print('negativo_x = ', x)

#Em c++, a utilização do operador é semelhante para soma, subtração, multiplicação, módulo e numero negativo. Não existe 
#no c++ a divisão mostrando um inteiro de resultado e nem potencia. 

#Operadores compostos 
print('Operadores compostos: ')
#Soma 
x += a
print('soma_x = ', x)
#Subtração
x -= a
print('subtração_x = ', x)
#Multiplicação
x *= a 
print('multiplicação_x = ', x)
#Divisão
x /= a  
print('divisão_x = ', x)
#Divisão que retorna apenas a parte inteira
x //= a  
print('divisão_x = ', x)
#Módulo
x %= a 
print('módulo_x = ', x)
#Potencia
x **= a 

print('------------------------------')
import math
fatorial = math.factorial(30)

print("O fatorial de 30 é:", fatorial)

#Em c++, o maior valor para um int de 32 bits é 2.147.483.647. O fatorial de trinta é um numero muito grande
#e o python consegue um muito muito mais longo 265252859812191058636308480000000.

print('------------------------------')

z = 5     #referência um inteiro
z1 = z    #referência o mesmo espaço na memória
print("z = ", z) 
print("z1 = ", z1) 

#Mas se eu incremento z, z1 ainda é 5
z += 1
print("z = ", z) 
print("z1 = ", z1) 


print('------------------------------')
print("métodos disponíveis para variáveis do tipo inteiro:")
print(dir(int))
```

```python
for i in range(10):
    char = chr(ord('0') + i)
    decimal = ord('0') + i
    print(f"'{char}' - Decimal: {decimal}")
    
for i in range(10):
    char = chr(ord('0') + i)
    decimal = ord('0') +i
    octal = oct(ord('0') +i)
    hexadecimal = hex(ord('0') +i)
    
    print(f"'{char}' - Decimal: {decimal}, 'Octal: '{octal}, 'Hexadecimal: '{hexadecimal}")
    
    
caractere = input("Digite um caractere: ")
decimal = ord(caractere)
octal = oct(ord(caractere))
hexadecimal = hex(ord(caractere))

print(f"Caractere digitado: '{caractere}'")
print(f"Decimal: {decimal}")
print(f"Octal: {octal}")
print(f"Hexadecimal: {hexadecimal}")
    
#e possivel trabalhar normalmente com caractere especial no python, isso por que são tratados como caracteres Unicode.
#sendo que o Unicode é um padrão de codificação que atribui um número único para cada caractere.
```

```python
nome_SobreNome = "Leane Soares Sousa"
nomeCompleto = nome_SobreNome.split()
nome = nomeCompleto[0]
sobrenome = ' '.join(nomeCompleto[1:])

print(nome)  
print(sobrenome)  


ordemAlfabetica = sorted([nome, sobrenome])
print("Ordem Alfabética:", ordemAlfabetica)

tamanhoNome = len(nome)
tamanhoSobrenome = len(sobrenome)

print("tamanho (Nome):", tamanhoNome)
print("tamanho (sobrenome):", tamanhoSobrenome)


palindromo = nome == nome[::-1]
print("É Palíndromo:", palindromo)
```
